# Replacing the outlier by median

```python
for cols in df.columns[:-1]:
  Q1 = df[cols].quantile(0.25)
  Q3 = df[cols].quantile(0.75)
  iqr = Q3 - Q1

  low = Q1-1.5*iqr
  high = Q3+1.5*iqr
  df.loc[(df[cols] < low) | (df[cols] > high), cols] = df[cols].median()
```

```python
x_test1 = pd.DataFrame(x_test_std,columns=["cement","slag","fly_ash","water","superplasticizer","coarse_agg","fine_aggregate","age"])
```
